```python
import numpy as np
import matplotlib.pyplot as plt

#Definimos la funciÃƒÂ³n (funciÃƒÂ³n objetiva)
def f(x):
    return np.sin(x)

# Definimos la derivada parcial respecto a x
def df_dx(x):
    return np.cos(x)

# interactive plot
def plot_f(current_pos):
    # Plot function
    x = np.arange(-5, 5, 0.1)
    y = f(x)
    plt.plot(x, y)
    plt.scatter(current_pos[0], current_pos[1], color='red')
   

def gradient_descen(iter, learing_rate, current_pos, break_iter):
    for i in range(iter):
        # gradient descent algorithm
        new_x = current_pos[0] - learing_rate * df_dx(current_pos[0])
        print(f'Iteraciones:{i} , Pendiente: {df_dx(current_pos[0])}')
        new_y = f(new_x)
        current_pos = (new_x, new_y)
        plot_f(current_pos)
        if break_iter:
            if df_dx(current_pos[0]) < 10e-10:
                break


iter = 1000  # Iteraciones
learing_rate = 0.1  # tasa de aprendizaje
current_pos = (0.8, f(0.8))  # Random Point
break_iter = True

# method
gradient_descen(iter, learing_rate, current_pos, break_iter)
```
